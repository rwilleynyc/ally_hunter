# Keras Initial Practice
## Data Preprocessing
### Generate Data

```python
# Create / Set Reproducibility
import numpy as np
import tensorflow as tf
import random as rn
from keras import backend as K
import os
import warnings

warnings.filterwarnings('ignore')

os.environ['PYTHONHASHSEED'] = '0'
np.random.seed(123)
rn.seed(123)
tf.set_random_seed(123)

session_conf = tf.ConfigProto(intra_op_parallelism_threads=1, inter_op_parallelism_threads=1)
sess = tf.Session(graph=tf.get_default_graph(), config=session_conf)
K.set_session(sess)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import warnings

warnings.filterwarnings('ignore')
```

Example data:

* An experimental drug was tested on individuals from ages 13 to 100
* The trial had 2,100 participants. Half were under 65 years old, and half were older
* 95% of patients 65 or older experienced side effects
* 95% of those younger had no side effects

```python
labels = []
samples = []

for i in range(2000):
    random_younger = np.random.randint(13, 64)
    samples.append(random_younger)
    n = 0 if (i % 100) != 0 else 1
    labels.append(n)
    
    random_older = np.random.randint(65, 100)
    samples.append(random_older)
    n = 1 if (i % 100) != 0 else 0
    labels.append(n)
    
labels = np.array(labels)
samples = np.array(samples)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_samples = scaler.fit_transform((samples).reshape(-1, 1)) # Reshape because fit_transform doesn't accept 1-d arrays
```

```python
train_samples, test_samples, train_labels, test_labels = train_test_split(scaled_samples, labels, test_size=.2, random_state=123)
```

```python
train_samples
```

## Model Initialization

```python
import keras
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

### Defining Network Structure

```python
model = Sequential([
    Dense(16, input_shape=(1,), activation = 'relu'),
    Dense(32, activation = 'relu'), 
    Dense(2, activation = 'softmax')
])
```

```python
model.summary()
```

### Compiling Model
* **Optimizer:** Typically **`Adam`** or Stochastic Gradient Descent (**`SGD`**)
* **Loss:** 'mean_squared_error' or others may be used
* **Metrics:** Typical scoring metrics available ('accuracy', 'recall', 'mae', etc.)

```python
model.compile(optimizer=Adam(lr=.0001), 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
```

## Fitting Model

```python
model.fit(train_samples, train_labels, 
          validation_split=.1,
          batch_size=10, 
          epochs=20, 
          shuffle=True, # Does not apply to validation set
          verbose=2) 
```

## Model Prediction

```python
predictions = model.predict(test_samples, 
                            batch_size=10)
```

```python
class_preds = model.predict_classes(test_samples, 
                                    batch_size=10)
```

```python
for i in range(len(predictions[:5])):
    age = '65+' if test_samples[i][0] >=.5 else '13+'
    print(f'Age: {age}')
    print(f'Prediction: {class_preds[i]}')
    print(f'Actual: {test_labels[i]}\n')
```

## Confusion Matrix

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import itertools
```

```python
cm = confusion_matrix(test_labels, class_preds)
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.figure(figsize=(10, 10))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
cm_plot_labels = ['No Side Effects', 'Has Side Effects']
plot_confusion_matrix(cm, cm_plot_labels, title='Confusion Matrix')
```

# Keras Convolutional Neural Network for Image Classifier

```python
from keras.layers.core import Flatten
from keras.preprocessing.image import ImageDataGenerator
from keras.layers.normalization import BatchNormalization
from keras.layers.convolutional import *
```

## Create Train, Validation, and Test Sets

```python
train_path = 'ally_and_hunter/train'
valid_path = 'ally_and_hunter/valid'
test_path = 'ally_and_hunter/test'
```

```python
train_batches = ImageDataGenerator().flow_from_directory(train_path, target_size=(224, 224), classes=['ally', 'hunter'], batch_size=9)
valid_batches = ImageDataGenerator().flow_from_directory(valid_path, target_size=(224, 224), classes=['ally', 'hunter'], batch_size=4)
test_batches = ImageDataGenerator().flow_from_directory(test_path, target_size=(224, 224), classes=['ally', 'hunter'], batch_size=23)
```

```python
def plots(ims, figsize=(12, 6), rows=1, interp=False, titles=None):
    if type(ims[0]) is np.ndarray:
        ims = np.array(ims).astype(np.uint8)
        if (ims.shape[-1] != 3):
            ims = ims.transpose((0, 2, 3, 1))
    f = plt.figure(figsize=figsize)
    cols = len(ims) // rows if len(ims) % 2 == 0 else len(ims) // rows + 1
    for i in range(len(ims)):
        sp = f.add_subplot(rows, cols, i+1)
        sp.axis('Off')
        if titles is not None:
            sp.set_title(titles[i], fontsize=16)
        plt.imshow(ims[i], interpolation=None if interp else 'none')
```

```python
imgs, labels = next(train_batches)
```

```python
plots(imgs, titles=labels)
```

## Initialize & Train Model

```python
model = Sequential([Conv2D(32, (3, 3), activation='relu', input_shape = (224, 224, 3)), 
                    Flatten(), 
                    Dense(2, activation='softmax')])
```

```python
model.compile(Adam(lr=.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit_generator(train_batches, steps_per_epoch=9, 
                    validation_data=valid_batches, validation_steps=3, epochs=5, verbose=2)
```

## Predict

```python
test_imgs, test_labels = next(test_batches)
plots(test_imgs, titles=test_labels)
```

```python
test_labels = test_labels[:,0]
test_labels
```

```python
predictions = model.predict_generator(test_batches, steps=1, verbose=2)
```

```python
predictions
```

```python
cm = confusion_matrix(test_labels, predictions[:,0])
```

```python
cm_plot_labels = ['Hunter', 'Ally']
plot_confusion_matrix(cm, cm_plot_labels, title='Confusion Matrix')
```

## Build Fine-Tuned VGG16 Model

```python
vgg16_model = keras.applications.vgg16.VGG16()
```

```python
vgg16_model.summary()
```

```python
type(vgg16_model)
```

### Convert VGG16 to Sequential Model

```python
model = Sequential()
for layer in vgg16_model.layers[:-1]: # Exclude predictions layer
    model.add(layer)
```

```python
model.summary()
```

```python
for layer in model.layers:
    layer.trainable = False
```

```python
model.add(Dense(2, activation='softmax'))
```

```python
model.summary()
```

## Train Fine-Tuned VGG16 Model

```python
model.compile(Adam(lr=.001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit_generator(train_batches, steps_per_epoch=9, 
                    validation_data=valid_batches, validation_steps=3, epochs=30, verbose=2)
```

```python
test_imgs, test_labels = next(test_batches)
plots(test_imgs, titles=test_labels)
```

```python
test_labels = test_labels[:,0]
test_labels
```

```python
predictions = model.predict_generator(test_batches, steps=1, verbose=0)
```

```python
predictions
```

```python
cm = confusion_matrix(test_labels, np.round(predictions[:, 0]))
```

```python
plot_confusion_matrix(cm, cm_plot_labels, title='Confusion Matrix')
```
